To identify the cluster with maximum number of deaths according to each of the category in the dataset

```python
global dataset

def Import():
    global dataset
    #To classify the 'Import' category
    dfImport = (dataset.loc[dataset['category'] == 'import'])
    dfImport
    #maximum value of deaths in the Import category
    maxImport = dfImport[['deaths']].max()
    maxImport
    #To identify which row has the maximum deaths in the Import category
    maxImport1 = dfImport.loc[dfImport['deaths'] == int(maxImport)]
    maxImport1
    return maxImport1

def Religious():
    #To classify the 'Religious' category
    dfReligious = (dataset.loc[dataset['category'] == 'religious'])
    dfReligious
    #maximum value of deaths in the Religious category
    maxRel = dfReligious[['deaths']].max()
    maxRel
    #To identify which row has the maximum deaths in the Religious category
    maxRel1 = dfReligious.loc[dfReligious['deaths'] == int(maxRel)]
    maxRel1
    return maxRel1

def Work():
    #To classify the 'Workplace' category
    dfWork = (dataset.loc[dataset['category'] == 'workplace'])
    dfWork
    #maximum value of deaths in the Workplace category
    maxWork = dfWork[['deaths']].max()
    maxWork
    #To identify which row has the maximum deaths in the Workplace category
    maxWork1 = dfWork.loc[dfWork['deaths'] == int(maxWork)]
    maxWork1
    return maxWork1

def Risk():
    #To classify the 'High Risk' category
    dfRisk = (dataset.loc[dataset['category'] == 'highRisk'])
    dfRisk
    #maximum value of deaths in the High Risk category
    maxRisk = dfRisk[['deaths']].max()
    maxRisk
    #To identify which row has the maximum deaths in the High Risk category
    maxRisk1 = dfRisk.loc[dfRisk['deaths'] == int(maxRisk)]
    maxRisk1
    return maxRisk1

def Community():
    #To classify the 'Community' category
    dfCommunity = (dataset.loc[dataset['category'] == 'community'])
    dfCommunity
    #maximum value of deaths in the community category
    maxCom = dfCommunity[['deaths']].max()
    maxCom
    #To identify which row has the maximum deaths in the community category
    maxCom1 = dfCommunity.loc[dfCommunity['deaths'] == int(maxCom)]
    maxCom1
    return maxCom1

def Detention():
    #To classify the 'Detention Centre' category
    dfDetention = (dataset.loc[dataset['category'] == 'detentionCentre'])
    dfDetention
    #maximum value of deaths in the detention centre category
    maxDet = dfDetention[['deaths']].max()
    maxDet
    #To identify which row has the maximum deaths in the detention centre category
    maxDet1 = dfDetention.loc[dfDetention['deaths'] == int(maxDet)]
    maxDet1
    return maxDet1

def Education():
    #To classify the 'Education' category
    dfEducation = (dataset.loc[dataset['category'] == 'education'])
    dfEducation
    #maximum value of deaths in the education category
    maxEdu = dfEducation[['deaths']].max()
    maxEdu
    #To identify which row has the maximum deaths in the education category
    maxEdu1 = dfEducation.loc[dfEducation['deaths'] == int(maxEdu)]
    maxEdu1
    return maxEdu1
```
