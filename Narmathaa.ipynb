# Distribution of Deaths according to each category

To view the Maximum number of Deaths in the original dataset

```python
#maximum value in the 'deaths' column
maxValues = dataset[['deaths']].max()
maxValues
#To identify which row has the maximum number of deaths
maxDeaths = dataset.loc[dataset['deaths'] == int(maxValues)]
maxDeaths
```
